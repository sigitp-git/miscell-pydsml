```python
# axes is array
```

```python
fig,axes = plt.subplots(nrows = 3, ncols = 1)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[0].set_title('Second Plot')

axes[2].plot(x,x)
axes[0].set_title('Third Plot')
```

```python
# plt.figure() have to manually add axes loc and size then access plot, set_title, label etc
```

```python
# plt.subplot() created fig and axes automatically and then call plot, title label etc. plt.tight_layout()
```

```python
# Figure Size and DPI (Dot Per Inch/Pixels per Inch)
```

```python
fig = plt.figure(figsize=(8,2),dpi=100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(figsize=(8,2),dpi=100)

axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2),dpi=100)

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')
ax.legend(loc=0)

ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('X')
```
