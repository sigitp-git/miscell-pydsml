```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# functional method of plotting
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'r-')
```

```python
plt.plot(x,y)
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object Oriented Method
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('x label')
axes.set_ylabel('y label')
axes.set_title('title')
```
